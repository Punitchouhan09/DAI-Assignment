```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# YouTube Video Data Analysis
# Load the dataset
file_path = "Most popular 1000 Youtube videos.csv"
df = pd.read_csv(file_path)

# Convert numerical columns from string to float (removing commas first)
columns_to_convert = ["Video views", "Likes", "Dislikes"]
for col in columns_to_convert:
    df[col] = df[col].astype(str).str.replace(",", "").astype(float)

# Drop rows where 'Category' is missing since it's categorical
df.dropna(subset=["Category"], inplace=True)

# Set plot style
sns.set_style("whitegrid")

# Univariate Analysis - Histograms
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(df["Video views"], bins=30, kde=True, ax=axes[0])
axes[0].set_title("Distribution of Video Views")
axes[0].set_xlabel("Video Views")

sns.histplot(df["Likes"], bins=30, kde=True, ax=axes[1])
axes[1].set_title("Distribution of Likes")
axes[1].set_xlabel("Likes")

sns.histplot(df["Dislikes"].dropna(), bins=30, kde=True, ax=axes[2])
axes[2].set_title("Distribution of Dislikes")
axes[2].set_xlabel("Dislikes")

plt.tight_layout()
plt.show()

# Additional Histogram for Category Distribution
plt.figure(figsize=(10, 5))
sns.countplot(y=df["Category"], order=df["Category"].value_counts().index)
plt.title("Distribution of Video Categories")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()

# Univariate Analysis - Boxplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.boxplot(y=df["Video views"], ax=axes[0])
axes[0].set_title("Boxplot of Video Views")
axes[0].set_ylabel("Video Views")

sns.boxplot(y=df["Likes"], ax=axes[1])
axes[1].set_title("Boxplot of Likes")
axes[1].set_ylabel("Likes")

sns.boxplot(y=df["Dislikes"].dropna(), ax=axes[2])
axes[2].set_title("Boxplot of Dislikes")
axes[2].set_ylabel("Dislikes")

plt.tight_layout()
plt.show()

# Multivariate Analysis - Scatter Plots
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.scatterplot(x=df["Video views"], y=df["Likes"], alpha=0.5, ax=axes[0])
axes[0].set_title("Video Views vs. Likes")
axes[0].set_xlabel("Video Views")
axes[0].set_ylabel("Likes")

sns.scatterplot(x=df["Video views"], y=df["Dislikes"], alpha=0.5, ax=axes[1])
axes[1].set_title("Video Views vs. Dislikes")
axes[1].set_xlabel("Video Views")
axes[1].set_ylabel("Dislikes")

plt.tight_layout()
plt.show()

# Multivariate Analysis - Category-wise Boxplots
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

sns.boxplot(x=df["Category"], y=df["Video views"], ax=axes[0])
axes[0].set_title("Video Views by Category")
axes[0].set_xlabel("Category")
axes[0].set_ylabel("Video Views")
axes[0].tick_params(axis='x', rotation=90)

sns.boxplot(x=df["Category"], y=df["Likes"], ax=axes[1])
axes[1].set_title("Likes by Category")
axes[1].set_xlabel("Category")
axes[1].set_ylabel("Likes")
axes[1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```
